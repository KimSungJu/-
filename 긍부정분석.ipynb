```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('appreply.csv')
df.head()
```

```python
df.dropna(inplace=True)
df.reset_index(inplace=True)
df.drop(['index','Unnamed: 0'],axis=1,inplace=True)
df = df.drop_duplicates(subset=['text'])
df.head()
```

```python
df.head()
```

```python
import re
from tqdm import tqdm
from konlpy.tag import Okt

noun_result = []

for temp in tqdm(df['text']):
    temp_result = re.sub(r'[^ㄱ-ㅣ가-힣\s]','',temp)
    okt = Okt()
    temp_nouns = okt.nouns(temp_result)
    temp_noins_result = [x for x in temp_nouns if len(x)>1]
    noun_result.append(temp_noins_result)

noun_result
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect = TfidfVectorizer(tokenizer=lambda x: x, lowercase=False)
tfidf = tfidf_vect.fit_transform(noun_result).toarray()
tfidf_vect.vocabulary_
tfidf.shape
tfidf
```

```python
df['score_real'] = np.where(df['score']>3,1,0)
df
```

```python
df['score_real'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(tfidf, df['score_real'], test_size=0.2, random_state=10, stratify=df['score_real'])
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state=10)
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_pred, y_test)
```
